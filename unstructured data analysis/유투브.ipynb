```python
cmt=driver.find_elements_by_css_selector('#content-text');cmt
```

```python
content=[]
for element in cmt:
    content.append(element.text)
```

```python
content=[]
for element in cmt:
    txt=element.text
    content.append(txt.lower())
```

```python
for i in range(0,len(content)):
    content[i]=content[i].replace('sims','')
    content[i]=content[i].replace('pack','')
    content[i]=content[i].replace('game','')
    content[i]=content[i].replace('pets','')
```

```python
content
```

```python
from sklearn.feature_extraction.text import CountVectorizer
import numpy
import operator
```

```python
cv = CountVectorizer(max_features=2000, stop_words='english')
```

```python
tdm = cv.fit_transform(content)
```

```python
words = cv.get_feature_names()
```

```python
count_mat = tdm.sum(axis=0)
```

```python
count = numpy.squeeze(numpy.asarray(count_mat))
```

```python
word_count = list(zip(words, count))
```

```python
word_count
```

```python
word_count = sorted(word_count,key=operator.itemgetter(1),reverse=True)
```

```python
word_count
```

```python
%matplotlib inline
from matplotlib import pyplot
from wordcloud import WordCloud
```

```python
wc = WordCloud(background_color='white', width=400, height=300)
```

```python
cloud = wc.fit_words(dict(word_count))
```
