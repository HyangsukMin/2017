```python
recipe=[]
for elm in root.cssselect('div.info_area > div.subject > strong > a:nth-child(1)'):
    print(elm.attrib['href'])
    res2=requests.get('http://terms.naver.com'+elm.attrib['href'])
    root2=lxml.html.fromstring(res2.text)
    cnt=root2.cssselect('#size_ct > p:nth-child(8)')
    if cnt==[] :
        next
    else :
        recipe.append(cnt[0].text_content())
```

```python
recipe=[]
for i in range(1,200):
    url='http://terms.naver.com/list.nhn?cid=48156&so=st1.dsc&categoryType=&categoryId=48156&viewType=&page={}'.format(i)
    res=requests.get(url)
    root=lxml.html.fromstring(res.text)
    for link in root.cssselect('div.info_area > div.subject > strong'):
        res2=requests.get('http://terms.naver.com'+link.cssselect('a')[0].attrib['href'])
        root2=lxml.html.fromstring(res2.text)
        cnt=root2.cssselect('#size_ct > p:nth-child(8)')
        if cnt==[]:
            next
        else :
            recipe.append(cnt[0].text_content())
```

```python
recipe
```

```python
recipe2=recipe
```

```python
for i in range(0,len(recipe2)):
    recipe2[i]=recipe2[i].replace('주재료','')
    recipe2[i]=recipe2[i].replace('부재료','')
    recipe2[i]=recipe2[i].replace('약간','')
    recipe2[i]=recipe2[i].replace('재료','')
    recipe2[i]=recipe2[i].replace('숟가락','')
```

```python
from konlpy.tag import Komoran
```

```python
tagger=Komoran()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
import numpy
import operator
```

```python
def get_noun(text):
    nouns = tagger.nouns(text)
    return [n for n in nouns if len(n) > 1]  # 2글자 이상인 명사만 추출
```

```python
cv = CountVectorizer(tokenizer=get_noun, max_features=300)
```

```python
tdm = cv.fit_transform(recipe2)
```

```python
words = cv.get_feature_names()
```

```python
count_mat = tdm.sum(axis=0)
```

```python
count = numpy.squeeze(numpy.asarray(count_mat))
```

```python
word_count = list(zip(words, count))
```

```python
word_count = sorted(word_count, key=operator.itemgetter(1), reverse=True)
```
